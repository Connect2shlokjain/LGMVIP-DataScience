## Initialization of Machine Learning with Training and testing and validate continued by ensembler using Random Forest Classifier

```python
X = data.drop(columns='Class')
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 15,criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

#### Summary: Our model is trained with 95.5% Accuracy from Iris data set using random forest algorithm.
